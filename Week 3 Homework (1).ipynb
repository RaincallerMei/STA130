````python
import pandas as pd
import plotly.graph_objects as go

# Load the dataset
pingees = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Filter out rows with missing flipper_length_mm values
pingees = pingees.dropna(subset=['flipper_length_mm'])

# Function to get the necessary statistics for marking the plot
def get_stats(data):
    mean = data.mean()
    median = data.median()
    flipper_min = data.min()
    flipper_max = data.max()
    std = data.std()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    
    return mean, median, flipper_min, flipper_max, q1, q3, std

# List of colors for each species
species_colors = {
    'Adelie': 'blue',
    'Chinstrap': 'green',
    'Gentoo': 'red'
}

# Create separate histograms for each species
for species in pingees['species'].unique():
    species_data = pingees[pingees['species'] == species]['flipper_length_mm']
    mean, median, flipper_min, flipper_max, q1, q3, std = get_stats(species_data)
    
    # Create a figure for each species
    fig = go.Figure()
    
    # Add histogram
    fig.add_trace(go.Histogram(x=species_data, nbinsx=30, marker_color=species_colors[species],
                               name=f'{species} Flipper Length'))

    # Add mean and median lines with legends
    fig.add_vline(x=mean, line=dict(color='green', width=2, dash='dash'), 
                  name=f'{species} Mean', showlegend=True)
    fig.add_vline(x=median, line=dict(color='blue', width=2, dash='dot'),
                  name=f'{species} Median', showlegend=True)
    
    # Add range (min to max) with a legend
    fig.add_vrect(x0=flipper_min, x1=flipper_max, fillcolor="yellow", opacity=0.2, line_width=0, 
                  name=f'{species} Range', showlegend=True)
    
    # Add IQR (Q1 to Q3) with a legend
    fig.add_vrect(x0=q1, x1=q3, fillcolor="orange", opacity=0.3, line_width=0,
                  name=f'{species} IQR', showlegend=True)
    
    # Add 2 standard deviations (mean - 2*std to mean + 2*std) with a legend
    fig.add_vrect(x0=mean - 2 * std, x1=mean + 2 * std, fillcolor="red", opacity=0.1, line_width=0, 
                  name=f'{species} 2 Std', showlegend=True)
    
    # Update layout with a title and labels
    fig.update_layout(
        title=f'Flipper Length Distribution for {species} Penguins',
        xaxis_title='Flipper Length (mm)',
        yaxis_title='Count',
        bargap=0.1,
        template='plotly_white',
        showlegend=True  # Make sure the legend is displayed
    )
    
    # Show the plot
    fig.show(renderer="png")
#     fig.show()
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
pingees = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Filter out rows with missing flipper_length_mm values
pingees = pingees.dropna(subset=['flipper_length_mm'])

# Function to get the necessary statistics for marking the plot
def get_stats(data):
    mean = data.mean()
    median = data.median()
    flipper_min = data.min()
    flipper_max = data.max()
    std = data.std()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    
    return mean, median, flipper_min, flipper_max, q1, q3, std

# Create a row of three plots (one for each species)
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Set a color palette
species_colors = {
    'Adelie': 'blue',
    'Chinstrap': 'green',
    'Gentoo': 'red'
}

# Loop through each species to create individual KDE plots
for ax, species in zip(axes, pingees['species'].unique()):
    species_data = pingees[pingees['species'] == species]['flipper_length_mm']
    mean, median, flipper_min, flipper_max, q1, q3, std = get_stats(species_data)
    
    # KDE plot for each species
    sns.kdeplot(species_data, ax=ax, fill=True, color=species_colors[species], label=f'{species} KDE')
    
    # Add mean and median lines
    ax.axvline(mean, color='green', linestyle='--', label='Mean')
    ax.axvline(median, color='blue', linestyle=':', label='Median')
    
    # Shade range (min to max)
    ax.axvspan(flipper_min, flipper_max, color='yellow', alpha=0.2, label='Range')
    
    # Shade IQR (Q1 to Q3)
    ax.axvspan(q1, q3, color='orange', alpha=0.3, label='IQR')
    
    # Shade 2 standard deviations (mean - 2*std to mean + 2*std)
    ax.axvspan(mean - 2 * std, mean + 2 * std, color='red', alpha=0.1, label='2 Std')
    
    # Titles and labels
    ax.set_title(f'{species} Penguins')
    ax.set_xlabel('Flipper Length (mm)')
    ax.set_ylabel('Density')

# Add a single legend across all plots
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=6)

# Adjust layout for clarity
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````

### Mean and Median:

    Mean (often referred to as the "average") is a measure of central tendency that represents the typical or expected value of a data set. It is calculated by summing all the data points and then dividing by the number of data points.
    
    Median is the middle value of a data set when the values are arranged in order. 


### Right and left skewness:

    Right skewness is when the mean is greater than the median. The distribution's long tail on the right (positive) side causes the mean to be pulled upward by a few higher values, but the median is more resilient to these extremes. This is the reason why this occurs.
    
    Left skewness is when the mean is lesser than the median. This happens because the distribution has a long tail on the left (negative) side, where a few lower values pull the mean downwards, but the median remains resistant to these lower extremes.


### What leads to the difference:

    Because the mean divides by the count after adding up all the data points, it is at risk of outliers and extreme values. The tail of a skewed distribution pushes the mean in its direction.

    The median is a more accurate measure of central tendency for skewed distributions because it is the middle value and is less impacted by extreme values.

````python
import pandas as pd
import plotly.express as px

# Load the dataset
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')

# Make identical boy and girl names distinct
bn['name'] = bn['name'] + " " + bn['sex'] 
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)
bn = bn.sort_values(['name', 'year'])

# Create the increase or decrease in name prevalence from the last year
bn['percent change'] = bn['percent'].diff()
new_name = [True] + list(bn.name[:-1].values != bn.name[1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent']
bn = bn.sort_values('year')

# Restrict to "common" names
bn = bn[bn.percent > 0.001]

# Create the animated scatter plot with specified parameters
fig = px.scatter(
    bn,
    x='percent change',  # Change in name popularity
    y='rank',            # Rank of the name
    animation_frame='year',  # Animation over years
    animation_group='name',   # Group by name
    size='percent',          # Size by prevalence
    color='sex',            # Differentiate by sex
    hover_name='name',      # Name in hover info
    size_max=50,
    range_x=[-0.005, 0.005]  # Set x-axis range
)

# Update y-axis to be reversed (rank 1 at the top)
fig.update_yaxes(autorange='reversed')

# Show the figure
fig.show(renderer="png")
````
